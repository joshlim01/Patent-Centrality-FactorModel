```python
import pandas as pd

# Provide the path to your Excel file
excel_file_path = 'CRSP_Data.csv'

# Load the Excel file into a pandas DataFrame
df = pd.read_csv(excel_file_path)

df.head()
```

```python
selected_columns = ['Global Company Key','Ticker Symbol', 'Company Name', 'Data Date','Market Value - Total']  # Replace with the actual column names

# Create a new DataFrame with only the selected columns
extracted_data = df[selected_columns]

extracted_data.head()

```

```python

# Provide the path to your Excel file
excel_file_path = 'Patent Data.csv'

# Load the Excel file into a pandas DataFrame
df = pd.read_csv(excel_file_path)

df.head()
```

```python
df1=extracted_data
df2=df
# Merge the dataframes based on the number key
merged_df = pd.merge(df1, df2, left_on='Global Company Key', right_on='gvkey')

merged_df.head()
```

```python
selected_columns = ['Ticker Symbol','Data Date', 'Market Value - Total', 'grant_year','patent_count']  # Replace with the actual column names

# Create a new DataFrame with only the selected columns
extracted_data = merged_df[selected_columns]

extracted_data.head()

```

```python
df=extracted_data
# Extract year from 'Data Date' column
df['Year'] = pd.to_datetime(df['Data Date']).dt.year

# Filter DataFrame where 'Year' matches 'grant_year'
filtered_df = df[df['Year'] == df['grant_year']]

filtered_df.head(20)
```

```python
df=filtered_df


# Divide Column1 by Column2 and store the result in a new column
df['Patent Intensity'] = df['patent_count']/df['Market Value - Total']
df.head()
```

```python

file_path = 'Patent_Intensity.xlsx'

# Save the DataFrame to Excel
df.to_excel(file_path, index=False)
```

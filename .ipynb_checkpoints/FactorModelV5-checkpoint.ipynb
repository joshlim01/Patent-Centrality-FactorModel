```python
import pandas as pd
import yfinance as yf
import statsmodels.formula.api as smf
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
RISKY_ASSET = 'AAPL' #INPUT STOCK HERE
START_DATE = '2010-12-31'
END_DATE = '2014-12-31'

#patent data does not exist for 2017
```

```python


# momentum factor
df_mom = web.DataReader('F-F_Momentum_Factor', 'famafrench', 
                        start=START_DATE)[0]
df_mom.index = df_mom.index.format()

# five factors
df_five_factor = web.DataReader('F-F_Research_Data_5_Factors_2x3', 
                                'famafrench', 
                                start=START_DATE)[0]
df_five_factor.index = df_five_factor.index.format()
df_five_factor
asset_df = yf.download(RISKY_ASSET, start=START_DATE, end=END_DATE, 
                       #auto_adjust=True
                       progress=False)
y = asset_df['Adj Close'].resample('M') \
                         .last() \
                         .pct_change() \
                         .dropna()

y.index = y.index.strftime('%Y-%m')
y.name = 'return'


# join all datasets on the index
five_factor_data = df_five_factor.join(y)


# rename columns
five_factor_data.columns = ['mkt', 'smb', 'hml', 'rmw', 'cma', 
                            'rf', 'rtn']

# divide everything (except returns) by 100
five_factor_data.loc[:, five_factor_data.columns != 'rtn'] /= 100

# select period of interest
five_factor_data = five_factor_data.loc[START_DATE:END_DATE]

# calculate excess returns
five_factor_data['excess_rtn'] = five_factor_data.rtn - five_factor_data.rf

five_factor_data.head()



# Specify the path to the Excel file
excel_file_path = "CentralityData.xlsx"

# Load the Excel file into a DataFrame
df = pd.read_excel(excel_file_path)

df = df.rename(columns={df.columns[0]: 'Date'})

df.head()

df['Date'] = pd.to_datetime(df['Date'])

# Extract year and month from the 'Date' column and create a new column
df['YearMonth'] = df['Date'].dt.strftime('%Y-%m')
df.head()

# Set 'YearMonth' column as the index
df = df.set_index('YearMonth')
df=df.drop(columns='Date',axis=1)

new_columns = {col: f'Cent_{col}' for col in df.columns}
df = df.rename(columns=new_columns)
df = df / 100
df.head()

df = pd.merge(five_factor_data, df, left_index=True, right_index=True, how='inner')





six_factor_data=df


Stock_OLS = 'Cent_' + RISKY_ASSET

```

```python
columns_to_keep = ['mkt','smb','hml','rmw','cma','rf','rtn','excess_rtn',Stock_OLS]  # Specify the columns you want to keep

six_factor_data = six_factor_data[columns_to_keep]
six_factor_data.head()
```

```python
# Specify the path to the Excel file
excel_file_path = "Patent_Intensity.xlsx"

# Load the Excel file into a DataFrame
df = pd.read_excel(excel_file_path)
df

df = df[df['Ticker Symbol'] == RISKY_ASSET]


df.tail()
columns_to_keep = ['Data Date','Patent Intensity']  # Specify the columns you want to keep

df = df[columns_to_keep]
df.head()

# Convert 'Data Date' column to datetime format
df['Data Date'] = pd.to_datetime(df['Data Date'])

# Set 'Data Date' as the index
df.set_index('Data Date', inplace=True)

# Convert index to year-month format
df.index = df.index.to_period('M')
df.head()

# Shift index by 1 month to create one-month prior DataFrame
one_month_prior = df.shift(1, freq='M')

# Shift index by -1 month to create one-month after DataFrame
one_month_after = df.shift(-1, freq='M')

# Concatenate the DataFrames
result = pd.concat([one_month_prior, df, one_month_after], axis=1)
result.columns = ['One Month Prior', 'Patent Intensity', 'One Month After']
result.head(50)

df=result
# Reset the index of the DataFrame
df_reset = df.reset_index()
df_reset

# Sort the DataFrame based on the 'Data Date' column
df_sorted = df_reset.sort_values('Data Date')
df_sorted.set_index('Data Date', inplace=True)

df_sorted
df=df_sorted
# Replace NaN values in 'Patent Intensity' column with non-NaN values from other columns
df['Patent Intensity'] = df['Patent Intensity'].fillna(df['One Month Prior'])
df['Patent Intensity'] = df['Patent Intensity'].fillna(df['One Month After'])

# Remove 'One Month Prior' and 'One Month After' columns
df.drop(['One Month Prior', 'One Month After'], axis=1, inplace=True)



df.head()
```

```python

df2=df
df1=six_factor_data

df1.index = pd.to_datetime(df1.index)
#df1.index = df1.index.to_timestamp()
df2.index = df2.index.to_timestamp()

#df2.index = pd.to_datetime(df2.index)

joined_df = df1.join(df2)

```

```python

joined_df.rename(columns={'Patent Intensity': 'PI'}, inplace=True)

joined_df.head()
df=joined_df
df

df.index = pd.to_datetime(df.index)

# Use the strftime() function to format the index column as "YYYY-MM"
df.index = df.index.strftime('%Y-%m')
df
```

```python
seven_factor_data=df

seven_factor_model = smf.ols(
    formula='excess_rtn ~ mkt + smb + hml + rmw + cma +' +Stock_OLS+ '+ PI',
    data=seven_factor_data
).fit()

print(seven_factor_model.summary())

```

```python
rf = seven_factor_data['rf'].mean()
market_premium = seven_factor_data['mkt'].mean()
smb_premium = seven_factor_data['smb'].mean()
hml_premium = seven_factor_data['hml'].mean()
rmw_premium=seven_factor_data['rmw'].mean()
cma_premium=seven_factor_data['cma'].mean()
Cent_premium=seven_factor_data[Stock_OLS].mean()
PI_premium=seven_factor_data['PI'].mean()

intercept, b1, b2, b3,b4,b5,b6,b7 = seven_factor_model.params

expected_monthly_return = rf + b1 * market_premium + b2 * smb_premium + b3 * hml_premium +b4 * rmw_premium + b5 * cma_premium +b6*Cent_premium+b7*PI_premium
expected_yearly_return = expected_monthly_return * 12
print("Expected yearly return:" + str(expected_yearly_return))
```

```python
num_rows = len(seven_factor_data)

print("Number of rows:", num_rows)
time=num_rows

categories = ['Market', 'Size', 'Value','Profitability ', 'Investment','Risk Free Rate','Centrality','Patent Intensity']
time_periods = seven_factor_data.index[:time].tolist()
data = np.array([seven_factor_data['mkt'].tolist()[:time],  # mkt
                 seven_factor_data['smb'].tolist()[:time],   # smb
                 seven_factor_data['hml'].tolist()[:time],
                 seven_factor_data['rmw'].tolist()[:time],   # Category 2
                 seven_factor_data['cma'].tolist()[:time],
                 seven_factor_data['rf'].tolist()[:time],   # cma
                 seven_factor_data[Stock_OLS].tolist()[:time],
                 seven_factor_data['PI'].tolist()[:time]
                 ])  

# Create stacked bar chart
fig, ax = plt.subplots()

# Plotting each category's bar
bottom = np.zeros(len(time_periods))
for i, category in enumerate(categories):
    ax.bar(time_periods, data[i], bottom=bottom, label=category)
    bottom += data[i]

# Customize the plot
ax.set_xlabel('Time')
ax.set_ylabel('Return')
ax.set_title(RISKY_ASSET+ ' Factor VS Actual Returns')
ax.legend()

# Calculate trend line
x = np.arange(len(time_periods))
y=five_factor_data['rtn'].tolist()[:time]

# Plot the trend line
ax.plot(x, y, color='black', label='Actual Return')
ax.legend()

# Set the legend below the graph
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=len(categories))

# Set the x-axis tick labels to display horizontally
plt.xticks(rotation=90)
plt.xticks(fontsize=8)
# Display the chart
plt.show()
```
